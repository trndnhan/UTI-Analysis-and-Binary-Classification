##Count the number of patients by gender and test result

```python
query = "SELECT Gender, Diagnosis, COUNT(*) AS Count FROM UTI GROUP BY Gender, Diagnosis;"
pd.read_sql(query, conn)
```

## Average age of patients with positive diagnosis

```python
query = "SELECT AVG(Age) FROM UTI WHERE Diagnosis = 'POSITIVE';"
pd.read_sql(query, conn)
```

```python
query = "SELECT Gender, AVG(Age) FROM UTI WHERE Diagnosis = 'POSITIVE' GROUP BY Gender;"
pd.read_sql(query, conn)
```

##Separate patients by age group

```python
query = """SELECT CASE
            WHEN Age < 18 THEN 'Children'
            WHEN Age >= 18 AND Age < 65 THEN 'Adult'
            ELSE 'Elder'
       END AS AgeGroup,
       COUNT(*) AS Count
FROM UTI
GROUP BY AgeGroup;"""
pd.read_sql(query, conn)
```

```python
query = """SELECT CASE
            WHEN Age < 18 THEN 'Children'
            WHEN Age >= 18 AND Age < 65 THEN 'Adult'
            ELSE 'Elder'
       END AS AgeGroup,
       COUNT(*) AS Count
FROM UTI
WHERE Diagnosis = 'POSITIVE'
GROUP BY AgeGroup;"""
pd.read_sql(query, conn)
```

```python
query = """SELECT CASE
            WHEN Age < 18 THEN 'Children'
            WHEN Age >= 18 AND Age < 65 THEN 'Adult'
            ELSE 'Elder'
       END AS AgeGroup, Gender,
       COUNT(*) AS Count
FROM UTI
GROUP BY AgeGroup, Gender;"""
pd.read_sql(query, conn)
```

```python
query = """SELECT CASE
            WHEN Age < 18 THEN 'Children'
            WHEN Age >= 18 AND Age < 65 THEN 'Adult'
            ELSE 'Elder'
       END AS AgeGroup, Gender,
       COUNT(*) AS Count
FROM UTI
WHERE Diagnosis = 'POSITIVE'
GROUP BY AgeGroup, Gender;"""
pd.read_sql(query, conn)
```
