# Data Pre-processing

```python
df.dropna(inplace=True)
```

```python
df.rename(columns=lambda x: x.replace(" ", "_"), inplace=True)
```

```python
df = df.drop(columns=['Unnamed:_0'])
```

```python
matched_values = df['WBC'].isin(df['RBC'])
matched_values_unique = df.loc[matched_values, 'WBC'].unique().tolist()
matched_values_unique
```

```python
unmatched_values_wbc = df.loc[~df['WBC'].isin(df['RBC']), 'WBC'].unique().tolist()
unmatched_values_wbc
```

```python
unmatched_values_rbc = df.loc[~df['RBC'].isin(df['WBC']), 'RBC'].unique().tolist()
unmatched_values_rbc
```

```python
def transform_wrbc(value):
    if '>' in value:
        return int(value.replace('>', ''))
    elif value == 'LOADED':
        return 150
    elif value == 'TNTC':
        return 200
    else:
        range_values = value.split('-')
        return (int(range_values[0]) + int(range_values[1])) / 2

df['WBC'] = df['WBC'].apply(transform_wrbc)
df['RBC'] = df['RBC'].apply(transform_wrbc)
```

```python
sorted(df['Age'].unique())[:13]
```

```python
def transform_age(age):
    if age < 1:
        return age * 100 / 12
    else:
        return age

df['Age'] = df['Age'].apply(transform_age)
```
